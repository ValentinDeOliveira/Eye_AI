```python
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
import glob
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
import random
from sys import stdout

listAcc = np.array([])
fauxPositif = np.array([])
fauxNegatif = np.array([])
```

```python
#########################################################################################################################
#                                                     ANALYSE                                                           #
#########################################################################################################################

listeImgOeilOuvert = list()
listeImgOeilFerme = list()

for i in glob.glob('mrlEyes_2018_01/ferme/*.png',recursive=True) :
    img = cv2.imread(i,0)
    img = cv2.resize(img, (100,100))
    img = img.reshape(img.shape[0]*img.shape[1])
    listeImgOeilOuvert.append(img)
    
for i in glob.glob('mrlEyes_2018_01/ouvert/*.png',recursive=True) :
    img = cv2.imread(i,0)
    img = cv2.resize(img, (100,100))
    img = img.reshape(img.shape[0]*img.shape[1])
    listeImgOeilFerme.append(img)
    
listeOeil = np.concatenate((listeImgOeilFerme, listeImgOeilOuvert), 0)

oeilFerme = np.c_[listeImgOeilFerme, np.zeros(len(listeImgOeilFerme))]
oeilOuvert = np.c_[listeImgOeilOuvert, np.ones(len(listeImgOeilOuvert))]

listeOeilFinal = np.concatenate((oeilOuvert,oeilFerme), 0)

listeImg2 = list()
for i in glob.glob('mrlEyes_2018_01/predictions/*.png',recursive=True):
    img = cv2.imread(i,0)
    img = cv2.resize(img, (100,100))
    img = img.reshape(img.shape[0]*img.shape[1])
    listeImg2.append(img)
    
listeImg2 = np.asarray(listeImg2)
print("Analyse des images terminée")
#########################################################################################################################
#                                                    PRÉDICTION                                                         #
#########################################################################################################################
    
acc = 0
repetition = 150
KNN = KNeighborsClassifier()
for i in range(0, repetition):
    stdout.write("\rrépétition : "+str(i+1)+"/"+str(repetition))
    X_train, X_test, y_train, y_test = train_test_split(listeOeil, listeOeilFinal[:,-1], test_size = 0.3,random_state=2*i)
    clf = KNN.fit(X_train, y_train)
    ypredit = clf.predict(X_test)
    
    listAcc =np.append(listAcc, accuracy_score(y_test, ypredit)) 
    tn, fp, fn, tp = metrics.confusion_matrix(y_test, ypredit).ravel()
    fauxPositif = np.append(fauxPositif,fp)
    fauxNegatif = np.append(fauxNegatif,fn)
    stdout.flush()

stdout.write("\r\r\n")
print("--------------------------------------------------------------------------------")
print("moyenne : "+ str(np.mean(listAcc)))
print("écart-type : "+str(np.var(listAcc)))
```

```python
#########################################################################################################################
#                                              AFFICHAGE IMAGES                                                         #
#########################################################################################################################


# /!\ LES CHIFFRES CORRESPONDENT A L'IMAGE D'EN DESSOUS /!\

X_test, y_test = train_test_split(listeImg2, test_size =1)
ypredit = clf.predict(X_test)

fig, axes = plt.subplots(6, 6, figsize=(12, 12),
subplot_kw={'xticks':[], 'yticks':[]},
gridspec_kw=dict(hspace=0.14, wspace=0))
print(len(ypredit))
for i, ax in enumerate(axes.flat):
    imgReshape = np.reshape(X_test[i], (-1, 100))
    ax.imshow(imgReshape, interpolation='nearest')
    ax.set_title(ypredit[i], y=-0.18)    
```

```python
#########################################################################################################################
#                                             ÉVOLUTION MOYENNE                                                         #
#########################################################################################################################
from matplotlib.pyplot import figure

figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')

donnees, = plt.plot(listAcc, 'r', label="test") 
moyenne, =plt.plot(np.full(listAcc.shape[0], np.mean(listAcc)), 'b', label="test2")

plt.legend([donnees,moyenne],["Précision", "Moyenne ≈ "+str(round(np.mean(listAcc),4))], fontsize=14)
plt.title("Évolution de la précision sur "+str(repetition)+" répétitions", fontsize=20)
plt.show()
```
